```python
import pandas as pd
import numpy as numpy
pd.set_option('display.max_columns', None)

import pickle as pkl
from sklearn.model_selection import train_test_split , GridSearchCV, PredefinedSplit
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
air_data = pd.read_csv('Invistico_Airline.csv')
air_data.head(10)
```

```python
air_data.dtypes
```

```python
air_data.info()
```

```python
air_data_subset = air_data.dropna(axis=0)
air_data_subset.info()
```

```python
air_data_subset.isnull().sum()
```

```python
air_data_subset = pd.get_dummies(air_data_subset, columns=['Customer Type', 'Type of Travel', 'Class'],dtype=int)
air_data_subset.head()
```

```python
air_data_subset.info()
```

```python
X = air_data_subset.drop('satisfaction', axis=1)
y = air_data_subset['satisfaction']
```

```python
# Separate into train, validate, test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.25,random_state=0)
```

```python
# Determine set of hyperparameters

cv_params = {'n_estimators':[50,100],
            'max_depth':[10,50],
            'min_samples_leaf':[0.5, 1],
            'min_samples_split':[0.001, 0.1],
            'max_features': ['sqrt'],
            'max_samples':[0.5, 0.9]}
```

```python
# Create list of split indices

split_index = [0 if x in X_val.index else -1 for x in X_train.index ]
custom_split = PredefinedSplit(split_index)
```

```python
# Instantiate model

rf = RandomForestClassifier(random_state=0)
```

```python
# Search over specified parameters

rf_val = GridSearchCV(rf, cv_params, cv = custom_split, refit='f1', n_jobs=-1, verbose=1 )
```

```python
%%time
# Fit the model.

rf_val.fit(X_train, y_train)
```

```python
# Obtain optimal parameters

rf_val.best_params_
```

```python
# Use optimal parameters on GridSearchCV

rf_opt = RandomForestClassifier(n_estimators = 50, max_depth= 50, min_samples_leaf =  1, 
             min_samples_split = 0.001, max_features = 'sqrt' , max_samples = 0.9 , random_state=0)
```

```python
# Fit the optimal model

rf_opt.fit(X_train, y_train)
```

```python
# Predict on test set

y_pred = rf_opt.predict(X_test)
```

```python
# Get precision score

pc_test = precision_score(y_test, y_pred, pos_label='satisfied') # pos_label: as positive class when calculating.
print('Precision: ','%0.3f'% pc_test)
print(f"which means of all positive predictions, {pc_test:0.1%} prediction are true positive.")
```

```python
# Get recall score

rc_test = recall_score(y_test, y_pred, pos_label='satisfied')
print('Recall: ','%0.3f'% rc_test)
print(f"which means of all real positive cases in test set, {rc_test:0.1%} are  predicted positive.")
```

```python
# Get accuracy score

ac_test = accuracy_score(y_test, y_pred)
print('Accuracy: ','%0.3f'% ac_test)
print(f"which means of all cases in test set, {ac_test:0.1%} are predicted true positive or true negative.")
```

```python
# Get F1 score

f1_test = f1_score(y_test, y_pred, pos_label='satisfied')
print('F1: ','%0.3f'% f1_test)
print(f"which means the test set's harmonic mean is {f1_test:.1%}")
```

```python
# Create table of results

table = pd.DataFrame({'Model':["Tuned Decision Tree", "Tuned Random Forest"],
                       'F1':  [0.945422, f1_test],
                        'Recall': [0.935863, rc_test],
                        'Precision': [0.955197, pc_test],
                        'Accuracy': [0.940864, ac_test]})
table
```
